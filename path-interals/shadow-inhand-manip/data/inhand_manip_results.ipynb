```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import os
```

```python
og_path = './percent_success/'
file_paths = [path for path in os.listdir(og_path) if path[0] != '.' and path != 'emppi' and path != 'unknown_param']
label_map = {
            'noupdate_param' : 'MPPI w/ imperfect model',
            'emppi2' : 'EMPPI',
            'known_param' : 'MPPI w/ perfect knowledge'
}
```

```python
plt.figure(figsize=(5.5,7))
ax = plt.subplot(2,1,1)

for path in file_paths:
    successes      = []
    time2Successes = []
    if path == 'emppi2':
        param_err      = []
    for i in range(1,6):
        _data = pickle.load(open(og_path + path + '/performance_data{}.pkl'.format(i), 'rb'))
        cma = 0.
        cma_hist = [cma]
        for i, d in enumerate(_data['successes']):
            cma = cma + (d - cma) / (i + 1)
            cma_hist.append(cma)
        attempts = np.arange(0, len(_data['successes'])) + 1
        _data['time2success'].insert(0, 0)
        _sampled_time = np.cumsum(_data['time2success'])
        _t_des = np.arange(0, 2000)
        _y_output = np.interp(_t_des, _sampled_time, cma_hist)
        time2Successes.append(_t_des)
        successes.append(_y_output)
        if path == 'emppi2':
            _errors = []
            for stats in _data['stats']:
                # true mass 0.07087500000000002
                _errors.append(
                    np.abs(stats['mean gain']-1)
                )

            param_err.append(np.stack(_errors)[:1000, :])
            
    succ_var = np.var(successes, axis=0)

#     time2Successes = np.mean(time2Successes, axis=0)
    successes      = np.mean(successes, axis=0)
    x_ax = np.arange(0, len(successes))
#     time2Successes /= np.max(time2Successes)
    ax.plot(time2Successes[0] * 5 * 0.002, successes, label=label_map[path])
#     ax.plot(successes, label=path)
    ax.fill_between(time2Successes[0] * 5 * 0.002, 
                    np.clip(successes-0.95*np.sqrt(succ_var), 0., np.inf), 
                    np.clip(successes+0.95*np.sqrt(succ_var), 0, 1), alpha=0.1)

ax.set_title('(a) Success Rate Over Time')
ax.set_ylabel('Success Rate (%)')
ax.set_xlabel('Time (s)')
ax.legend(framealpha=0.)

ax = plt.subplot(2,1,2)

mean_err = np.mean(param_err, axis=0)
mean_err /= np.max(mean_err, axis=0)

_time = np.arange(0, len(mean_err)) * 5 * 0.002

num_param = mean_err.shape[1]
for i in range(num_param):
    ax.plot(_time, mean_err[:,i], label='P gain{}'.format(i))

ax.legend(framealpha=0.,  ncol=2, loc=(0.45,0.28))

# ax.legend(framealpha=0., loc=(0.6,0.7))
ax.set_ylabel('Normalized Absolute Error')
ax.set_title('(b) Proportional Actuator Gain Error')
ax.set_xlabel('Time (s)')

plt.tight_layout()

plt.savefig('method_comparison_inhandmanip.pdf', bbox_inches='tight')

# plt.show()
```
